```python
# Calculating the proportions by grouping and normalizing
proportions = data.groupby(['How likely are you to use the presented functionality?', 'What kind of neurodevelopmental disorder/s do you have? ']).size() / data.groupby(['How likely are you to use the presented functionality?']).size()
proportions
```

```python
# Setting the proportion index
proportions = proportions.reset_index(name='Proportion')
```

```python
# a normalized bar plot
plt.figure(figsize=(8, 6))
sns.barplot(data=proportions, x='How likely are you to use the presented functionality?', y='Proportion', hue='What kind of neurodevelopmental disorder/s do you have? ', palette='Set2')
plt.xlabel('How likely are you to use the presented functionality?')
plt.ylabel('Proportion')
plt.title('Normalized Bar Plot by Category and Variable')
plt.legend(title='What kind of neurodevelopmental disorder/s do you have? ', loc='upper right')
plt.show()
```

```python
# Perform a chi-squared test
observed = data.groupby(['How likely are you to use the presented functionality?', 'What kind of neurodevelopmental disorder/s do you have? ']).size().unstack()
chi2, p, _, _ = chi2_contingency(observed)
print(f'Chi-squared statistic: {chi2:.2f}')
print(f'p-value: {p:.4f}')
```

```python
import pandas as pd
from imblearn.under_sampling import RandomUnderSampler
from sklearn.preprocessing import MinMaxScaler
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd 
from scipy.stats import chi2_contingency
```

```python
data = pd.read_csv("C:/Users/Admin/Desktop/Neruodivergent project/Cleaned_data_ND.csv")
```

```python
data
```

```python
def match_phrases(text):
        phrases_to_match = ['spectrum disorder', 'deficit hyperactivity disorder']
        
        found_phrases = []
        
        for phrase in phrases_to_match:
            if phrase in text.lower():
                found_phrases.append(phrase)
                
                if len(found_phrases) == 0:
                    return 'Others'
                elif len(found_phrases) == 1:
                    if found_phrases[0] == 'spectrum disorder':
                        return 'Autism'
                    elif found_phrases[0] == 'deficit hyperactivity disorder':
                        return 'ADHD'
                    else: 
                        return 'Both'
        
```

```python
data['What kind of neurodevelopmental disorder/s do you have? '] = data['What kind of neurodevelopmental disorder/s do you have? '].apply(match_phrases)
                        
```

```python
# Replace 'None' values with 'Others' in a specific column
column_name = 'What kind of neurodevelopmental disorder/s do you have? '
data[column_name].fillna('Others', inplace=True)
```

# Scatter Plot 

```python
# Splitting the data into two groups: Autism and Others
autism_data = data[data['What kind of neurodevelopmental disorder/s do you have? '] == 'Autism']
ADHD_data = data[data['What kind of neurodevelopmental disorder/s do you have? '] == 'ADHD']
others_data = data[data['What kind of neurodevelopmental disorder/s do you have? '] == 'Others']
```

```python
autism_data
```

```python
# Creating a pivot table to count data points at each position
autism_data = pd.pivot_table(autism_data, index=['How knowledgeable are you about artificial intelligence?', 'How likely are you to use the presented functionality?'], values='What kind of neurodevelopmental disorder/s do you have? ', aggfunc='count')

# Reseting the index and rename the columns for better readability
autism_data = autism_data.reset_index()
autism_data = autism_data.rename(columns={'What kind of neurodevelopmental disorder/s do you have?': 'Count'})
autism_data
```

```python
# Creating the scatter plot with marker sizes based on the count
plt.figure(figsize=(20, 12))
scaling_factor = 80
plt.scatter(
    autism_data['How knowledgeable are you about artificial intelligence?'],
    autism_data['How likely are you to use the presented functionality?'],
    s = autism_data['What kind of neurodevelopmental disorder/s do you have? '] * scaling_factor,  # Use the count as marker size
    c = 'blue'
)

# Setting axis labels and a title
plt.xlabel('AI Knowledge')
plt.ylabel('Likelihood to Use Functionality')
plt.title('Scatter Plot with Marker Size Based on Data Point Count')

plt.show()
```
